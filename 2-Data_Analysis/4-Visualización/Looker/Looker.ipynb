**Paso 0: Preparación y Entendimiento Inicial**
Antes de sumergirnos en Looker Studio, es crucial entender qué vamos a construir. El objetivo es crear un dashboard interactivo que visualice datos de pedidos de una empresa.

**Paso 1: Investigación de Datos**
1.  **Fuente de Datos:** Vemos que los datos residen en un archivo de **Google Sheets**. Esta es una fuente común y fácil de conectar con Looker Studio.
2.  **Contenido de los Datos:** Los datos representan información de pedidos de una empresa ficticia. Columnas:
    *   `Row ID` (ID de Fila)
    *   `Order ID` (ID de Pedido)
    *   `Order Date` (Fecha de Pedido)
    *   `Ship Date` (Fecha de Envío)
    *   `Ship Mode` (Modo de Envío)
    *   `Customer ID` (ID de Cliente)
    *   `Customer Name` (Nombre del Cliente)
    *   `Segment` (Segmento: Consumidor, Corporativo, Oficina en Casa)
    *   `Country` (País)
    *   `City` (Ciudad)
    *   `State` (Estado)
    *   `Postal Code` (Código Postal)
    *   `Region` (Región)
    *   `Product ID` (ID de Producto)
    *   `Category` (Categoría: Tecnología, Material de Oficina, Mobiliario)
    *   `Sub-Category` (Subcategoría)
    *   `Product Name` (Nombre del Producto)
    *   `Sales` (Ventas)
    *   `Quantity` (Cantidad)
    *   `Discount` (Descuento)
    *   `Profit` (Beneficio)
    El video menciona que el conjunto de datos tiene 21 columnas y aproximadamente 9,994 filas.
3.  **Importancia:** Es fundamental comprender la estructura y el contenido de nuestros datos antes de visualizarlos. Esto informará las decisiones de diseño y los tipos de gráficos que crearemos para responder a preguntas de negocio.

**Paso 2: Conectar Looker Studio a Google Sheets**
1.  **Acceder a Looker Studio:** Abrimos Looker Studio (anteriormente Google Data Studio) yendo a `lookerstudio.google.com`. Probablemente empezaremos con un informe en blanco o crearemos uno nuevo.
2.  **Añadir Fuente de Datos:**
    *   En el menú superior, hacemos clic en **"Recurso"**.
    *   Seleccionamos **"Gestionar fuentes de datos añadidas"**.
    *   Hacemos clic en el botón **"+ AÑADIR UNA FUENTE DE DATOS"**.
    *   De la lista de conectores de Google, elegimos **"Hojas de cálculo de Google"**. Esto nos permite usar datos directamente desde nuestros archivos de Sheets.
    *   Buscamos y seleccionamos la hoja de cálculo que contiene nuestros datos (en el video, se llama "Data Source - Test").
    *   Nos aseguramos de que la hoja de trabajo correcta esté seleccionada (generalmente "Sheet1" o "Hoja1").
    *   Verificamos que las opciones **"Usar la primera fila como encabezados"** e **"Incluir celdas ocultas y filtradas"** estén marcadas. La primera es vital para que Looker Studio entienda los nombres de nuestras columnas.
    *   Hacemos clic en **"AÑADIR"** en la esquina inferior derecha.
    *   Aparecerá un mensaje de confirmación; hacemos clic en **"AÑADIR AL INFORME"**.
    *   **Por qué:** Este paso establece la conexión entre nuestro informe de Looker Studio y la fuente de datos en Google Sheets, permitiendo que los datos fluyan y se utilicen para crear visualizaciones.

**Paso 3: Establecer un Plan de Acción**
1.  **Visión del Dashboard:** Antes de empezar a arrastrar y soltar elementos, es útil tener una idea clara de lo que queremos mostrar. Arriba se muestra el dashboard final como referencia. Nuestro objetivo es crear un dashboard de pedidos que sea interactivo y muestre métricas clave.
2.  **Componentes Planeados:**
    *   **Filtros:** Para Región, Categoría, Segmento y Modo de Envío. Esto permitirá a los usuarios explorar los datos de forma dinámica.
    *   **Indicadores Clave de Rendimiento (KPIs):** Número total de pedidos, número de clientes, número de estados y número de categorías. Estos ofrecen un resumen rápido de las métricas más importantes.
    *   **Gráficos de Donut:** Para visualizar pedidos por Región, Categoría y Segmento, mostrando proporciones.
    *   **Gráfico de Barras:** Para comparar pedidos por ciudad.
    *   **Gráficos de Líneas:** Para mostrar tendencias de pedidos a lo largo del tiempo (anual y mensual).
    *   **Tablas:** Para mostrar detalles específicos de los pedidos.
    *   **Por qué:** Tener un plan nos ayuda a organizar el lienzo y asegurar que incluimos toda la información relevante de manera coherente.

**Paso 4: Diseño del Dashboard - Fondo y Formas Base**
1.  **Fondo del Dashboard:**
    *   Hacemos clic en **"Tema y diseño"** en la barra de herramientas superior.
    *   En el panel derecho que aparece, en la pestaña **"TEMA"**, hacemos clic en **"Personalizar"**.
    *   En la sección **"Fondo del informe"**, seleccionamos un color. Usemos negro, lo que hace que los colores brillantes de los gráficos resalten.
2.  **Forma del Encabezado (Header):**
    *   Insertamos una forma rectangular para el encabezado. Hacemos clic en el icono de **"Forma"** (un cuadrado y un círculo superpuestos) en la barra de herramientas y seleccionamos **"Rectángulo"**.
    *   Dibujamos el rectángulo en la parte superior del lienzo, ajustando su tamaño.
    *   En el panel de **"Propiedades del rectángulo"** a la derecha, vamos a "Fondo y borde". Cambiamos el color de fondo (el video usa un gris oscuro con baja opacidad, por ejemplo, un 10% de opacidad para que se vea sutilmente sobre el fondo negro).
3.  **Estructura con Formas (Layout):**
    *   Copiamos (Ctrl+C o Cmd+C) y pegamos (Ctrl+V o Cmd+V) el rectángulo del encabezado varias veces.
    *   Redimensionamos y reposicionamos estos rectángulos copiados para crear las diferentes secciones donde irán los KPIs, filtros y gráficos. Esto ayuda a organizar visualmente el dashboard. El video crea:
        *   4 rectángulos pequeños en la parte superior para KPIs.
        *   Un rectángulo vertical a la izquierda para los filtros.
        *   Varios rectángulos de diferentes tamaños en el área principal para los gráficos.
    *   **Por qué:** Usar formas como contenedores visuales ayuda a delimitar las diferentes secciones del dashboard, mejorando la legibilidad y la estética general. La opacidad sutil evita que las formas distraigan de los datos.

**Paso 5: Creación de Títulos**
1.  **Título Principal del Dashboard:**
    *   Insertamos un cuadro de texto haciendo clic en el icono **"Texto"** (la letra "T" mayúscula) en la barra de herramientas.
    *   Lo colocamos sobre la forma del encabezado.
    *   Escribimos el título principal, por ejemplo, "ORDERS DASHBOARD" (o "INFORME DE PEDIDOS").
    *   En el panel de **"Propiedades del texto"** a la derecha, personalizamos:
        *   Fuente: El video usa "Orbitron".
        *   Tamaño: Ej. 40px.
        *   Color: Ej. un magenta o rosa brillante.
        *   Alineación: Centrado.
2.  **Títulos para Gráficos y Secciones:**
    *   Copiamos y pegamos el cuadro de texto del título principal.
    *   Movemos estas copias a cada una de las secciones rectangulares que creamos para los gráficos.
    *   Editamos el texto para reflejar el contenido de cada sección (ej. "# of Orders by Region", "Orders - Trend Year", "# of Orders by City", etc.).
    *   Ajustamos el estilo de estos subtítulos:
        *   Fuente: Puede ser la misma ("Orbitron") o una más estándar.
        *   Tamaño: Más pequeño, ej. 12-14px.
        *   Color: Un color que contraste pero sea menos prominente que el título principal (ej. un azul claro o cian).
    *   **Por qué:** Los títulos son esenciales para que los usuarios entiendan rápidamente qué información se presenta en cada parte del dashboard.

**Paso 6: Creación de Filtros Interactivos**
1.  **Filtro de Región (Lista Desplegable):**
    *   Hacemos clic en **"Añadir un control"** en la barra de herramientas y seleccionamos **"Lista desplegable"**.
    *   Lo colocamos en la sección de filtros (el rectángulo vertical a la izquierda).
    *   En el panel de **"Configuración"** (pestaña "DATOS") a la derecha, en "Campo de control", arrastramos el campo **"Region"** desde nuestra lista de campos de la fuente de datos.
    *   En la pestaña **"ESTILO"**, ajustamos los colores del texto del encabezado y las opciones para que coincidan con el diseño.
2.  **Filtros de Segmento, Categoría y Modo de Envío (Lista de Tamaño Fijo):**
    *   Repetimos el proceso de "Añadir un control", pero esta vez seleccionamos **"Lista de tamaño fijo"** para cada uno de estos filtros.
    *   Los colocamos debajo del filtro de Región.
    *   Para cada uno, arrastramos el campo correspondiente ("Segment", "Category", "Ship Mode") al área "Campo de control".
    *   En "ESTILO", personalizamos los colores del texto, las casillas de verificación (checkboxes) y el fondo para que sean coherentes con el diseño oscuro del dashboard (ej. texto claro, color de acento para las casillas).
3.  **Líneas Separadoras entre Filtros:**
    *   Insertamos "Línea" (desde el icono de formas) entre cada filtro.
    *   Ajustamos su color (ej. el mismo rosa/magenta usado en los títulos) y grosor.
    *   **Por qué:** Los filtros permiten a los usuarios interactuar con el dashboard y ver subconjuntos de datos específicos. Las listas desplegables son buenas para campos con muchas opciones, mientras que las listas de tamaño fijo son mejores para menos opciones donde se pueden ver todas a la vez. Las líneas ayudan a la organización visual.

**Paso 7: Creación de KPIs (Indicadores Clave de Rendimiento)**
1.  **Iconos para KPIs:**
    *   Hacemos clic en el icono de **"Imagen"** en la barra de herramientas y seleccionamos **"Subir del ordenador"**.
    *   Subimos y colocamos los iconos apropiados (un camión para pedidos, un grupo de personas para clientes, un marcador de mapa para estados, una lista para categorías) en cada uno de los 4 rectángulos superiores que preparamos. Redimensionamos según sea necesario.
2.  **KPI de Número de Pedidos:**
    *   Hacemos clic en **"Añadir un gráfico"** y seleccionamos **"Tarjeta de resultados"** (Scorecard).
    *   La colocamos junto al icono del camión.
    *   En el panel de **"Configuración"** (pestaña "DATOS"):
        *   Métrica: Arrastramos **"Order ID"**. Por defecto, Looker Studio podría intentar sumarlo. Hacemos clic en el lápiz junto a "Order ID" (o donde dice "SUM") y cambiamos la agregación a **"Count Distinct" (Contar distintos)** para obtener el número único de pedidos.
        *   Podemos renombrar la métrica haciendo clic en el lápiz y escribiendo, por ejemplo, "# of Orders".
    *   En la pestaña **"ESTILO"**:
        *   Ocultamos el nombre de la métrica (si ya tenemos un título de texto encima).
        *   Centramos el valor numérico.
        *   Ajustamos el color y tamaño del texto del valor.
        *   Desmarcamos "Números compactos" si queremos ver el número completo en lugar de "10K".
3.  **Otros KPIs:**
    *   Copiamos y pegamos la tarjeta de resultados creada.
    *   Para el KPI **"# of Customers"**: Cambiamos la Métrica a **"Customer ID"** (Count Distinct). Renombramos.
    *   Para el KPI **"# of States"**: Cambiamos la Métrica a **"State"** (Count Distinct). Renombramos.
    *   Para el KPI **"# of Categories"**: Cambiamos la Métrica a **"Category"** (Count Distinct). Renombramos. (El video usa "Segment" para el último KPI, lo cual es un error si el icono es una lista para "Categorías"; el dashboard final sí muestra "Categories").
    *   Ajustamos el estilo de cada uno para que sea consistente.
    *   **Por qué:** Los KPIs proporcionan resúmenes numéricos de alto nivel de las métricas más importantes. Los iconos ayudan a identificar rápidamente qué representa cada KPI.

**Paso 8: Creación de Gráficos de Donut**
1.  **Pedidos por Región:**
    *   Hacemos clic en **"Añadir un gráfico"** y seleccionamos **"Gráfico de anillos"** (Donut chart).
    *   Lo colocamos en la primera sección designada para estos gráficos.
    *   En **"Configuración"** (pestaña "DATOS"):
        *   Dimensión: Arrastramos **"Region"**.
        *   Métrica: Arrastramos **"Order ID"** (y nos aseguramos que la agregación sea "Count" o "Count Distinct", según la necesidad; el video parece usar "Count" aquí).
    *   En **"ESTILO"**:
        *   En "Color por", seleccionamos **"Color único"** y elegimos un color (ej. el rosa/magenta).
        *   En "Leyenda", seleccionamos "Ninguna" para ocultarla, ya que el título del gráfico ya indica qué se muestra.
2.  **Pedidos por Categoría y Pedidos por Segmento:**
    *   Copiamos y pegamos el gráfico de donut dos veces.
    *   Para el segundo gráfico (Pedidos por Categoría): Cambiamos la Dimensión a **"Category"**.
    *   Para el tercer gráfico (Pedidos por Segmento): Cambiamos la Dimensión a **"Segment"**.
    *   Ajustamos los colores si es necesario para mantener la coherencia visual.
    *   **Por qué:** Los gráficos de donut son excelentes para mostrar la proporción de un total que corresponde a diferentes categorías de una dimensión.

**Paso 9: Creación de Gráfico de Barras (Pedidos por Ciudad)**
1.  **Insertar Gráfico de Barras:**
    *   Hacemos clic en **"Añadir un gráfico"** y seleccionamos **"Gráfico de barras"**.
    *   Lo colocamos en su sección designada.
    *   En **"Configuración"** (pestaña "DATOS"):
        *   Dimensión: Arrastramos **"City"**.
        *   Métrica: Arrastramos **"Order ID"** (agregación "Count").
    *   En **"ESTILO"**:
        *   En la sección de "Gráfico de barras", podemos limitar el número de barras a mostrar (ej. 5 o 10).
        *   Cambiamos el color de las barras (ej. rosa/magenta).
        *   En "Cuadrícula", podemos cambiar el color de las líneas de la cuadrícula a transparente para eliminarlas.
        *   En "Leyenda", seleccionamos "Ninguna".
        *   Ajustamos colores de etiquetas de los ejes si es necesario.
    *   **Por qué:** Los gráficos de barras son útiles para comparar valores entre diferentes elementos de una categoría (en este caso, ciudades).

**Paso 10: Creación de Gráficos de Líneas (Tendencias)**
1.  **Tendencia de Pedidos (Anual):**
    *   Hacemos clic en **"Añadir un gráfico"** y seleccionamos **"Gráfico de líneas"**.
    *   Lo colocamos en la sección "Orders - Trend Year".
    *   En **"Configuración"** (pestaña "DATOS"):
        *   Dimensión: Arrastramos **"Order Date"**. Hacemos clic en el icono de calendario junto a "Order Date", vamos a "Tipo" -> "Fecha y Hora" -> **"Año"**.
        *   Métrica: Arrastramos **"Quantity"** (Cantidad).
    *   En **"ESTILO"**:
        *   Cambiamos el color de la línea (ej. rosa/magenta).
        *   Activamos **"Mostrar puntos"**.
        *   En "Cuadrícula", color transparente.
        *   En "Leyenda", "Ninguna".
        *   En "Línea de referencia", hacemos clic en "Añadir una línea de referencia". Tipo: Métrica, Métrica: Quantity, Cálculo: Promedio. Esto mostrará una línea con el promedio de cantidad.
2.  **Tendencia de Pedidos (Mensual):**
    *   Copiamos y pegamos el gráfico de línea anual. Lo colocamos en la sección "Orders - Trend Month".
    *   En **"Configuración"** (pestaña "DATOS"):
        *   Para la Dimensión **"Order Date (Year)"**, hacemos clic en el lápiz, vamos a "Tipo" -> "Fecha y Hora" -> **"Mes"**.
    *   Ajustamos el estilo si es necesario (el video mantiene el mismo estilo).
    *   **Por qué:** Los gráficos de líneas son ideales para mostrar tendencias a lo largo del tiempo, permitiendo ver patrones, crecimiento o declive.

**Paso 11: Creación de Tablas de Detalles**
1.  **Tabla 1 (Detalles de Pedido):**
    *   Hacemos clic en **"Añadir un gráfico"** y seleccionamos **"Tabla"**.
    *   La colocamos en la primera sección designada para tablas en la parte inferior.
    *   En **"Configuración"** (pestaña "DATOS"):
        *   Dimensiones: Arrastramos **"Order Date"**, **"Order ID"**.
        *   Métrica: Arrastramos **"Quantity"**.
    *   En **"ESTILO"**:
        *   En "Colores de la tabla", ajustamos el color del texto del encabezado y de las celdas (ej. texto claro).
        *   En "Cuerpo de la tabla", desmarcamos **"Números de fila"**.
        *   En "Pie de tabla", desmarcamos **"Mostrar paginación"**.
        *   Alineamos el texto de las columnas (ej. centrado o a la izquierda según preferencia).
        *   Podemos hacer que las barras de las métricas sean de un color específico (ej. rosa).
2.  **Tabla 2 (Cantidad por Estado/Región):**
    *   Copiamos y pegamos la Tabla 1. La colocamos a su derecha.
    *   En **"Configuración"** (pestaña "DATOS"):
        *   Quitamos las dimensiones "Order Date" y "Order ID".
        *   Añadimos como Dimensiones: **"State"**, **"Region"**.
        *   Mantenemos la Métrica: **"Quantity"**.
3.  **Tabla 3 (Cantidad por Cliente):**
    *   Copiamos y pegamos la Tabla 2. La colocamos a su derecha.
    *   En **"Configuración"** (pestaña "DATOS"):
        *   Quitamos las dimensiones "State" y "Region".
        *   Añadimos como Dimensión: **"Customer ID"**. (El video parece añadir "Order ID" y "Quantity" de nuevo, pero el dashboard final muestra "Customer ID" y "Quantity").
        *   Mantenemos la Métrica: **"Quantity"**.
    *   Ajustamos el estilo de cada tabla para que sea consistente y legible.
    *   **Por qué:** Las tablas son útiles para mostrar datos detallados y específicos que pueden no ser evidentes en los gráficos agregados. Permiten una exploración más granular.

**Paso 12: Revisión Final y Conclusión**
1.  **Revisión General:** Observamos el dashboard completo. Verificamos que todos los elementos estén alineados, los títulos sean correctos y los filtros funcionen como se espera.
2.  **Interactividad:** Probamos los filtros para asegurarnos de que los gráficos y KPIs se actualizan correctamente.
3.  **¡Felicidades!** Hemos completado la creación de un dashboard funcional y visualmente atractivo en Looker Studio. Este dashboard ahora puede usarse para analizar el rendimiento de los pedidos y compartir información clave con los interesados.

```python
df = pd.read_csv("DataSource.csv")

# --- Step 1: Clean column names ---
# Convert to lowercase
df.columns = df.columns.str.lower()
# Replace spaces with underscores
df.columns = df.columns.str.replace(' ', '_')
# Remove special characters (you can customize this regex)
# This regex keeps letters, numbers, and underscores.
df.columns = df.columns.str.replace(r'[^a-z0-9_]', '', regex=True)
# Remove leading/trailing underscores or double underscores
df.columns = df.columns.str.strip('_')
df.columns = df.columns.str.replace('__', '_')
df.columns = df.columns.str.replace('-', '_')

# Optional: Check for duplicate column names after cleaning
if len(df.columns) != len(set(df.columns)):
    print("Warning: Duplicate column names found after cleaning. You might need to rename them manually or add suffixes.")
    # Example to handle duplicates (add _1, _2, etc.)
    cols = pd.Series(df.columns)
    for dup in cols[cols.duplicated()].unique():
        cols[cols[cols == dup].index.values.tolist()] = [dup + '_' + str(i) if i != 0 else dup for i, iota in enumerate(cols[cols == dup].index.values.tolist())]
    df.columns = cols


# --- Step 2: Use a clean table name ---
table_name = "data_source_table"
df.to_sql(table_name, con=engine, if_exists="replace", index=False) # index=False avoids writing DataFrame index as a column

# --- Step 3: Query using the clean table name ---
query = f"""
SELECT * FROM "{table_name}"
"""
# If the table name is simple and lowercase, you might not need quotes:
# query = f"""
# SELECT * FROM {table_name}
# """

result_df = pd.read_sql_query(query, con=engine)
print(result_df.head())
```
