```python
import requests
import zipfile
import os
import json
import pandas as pd
import numpy as np
```

```python
url = "https://valencia.opendatasoft.com/api/explore/v2.1/catalog/datasets/centros-educativos-en-valencia/records?limit=40"

api = requests.get(url)


data = api.json()
df_centros_edu = pd.DataFrame(data["results"])

# Ahora df_centros_edu es un DataFrame de Pandas donde cada fila
# representa un registro del JSON (un centro educativo).

# Puedes iterar sobre las filas del DataFrame usando .iterrows():
# for index, row in df_centros_edu.iterrows():
#     # Cada 'row' es una Serie de Pandas que representa una fila
#     # Puedes acceder a los datos por nombre de columna (clave del JSON)
#     nombre_centro = row["record"]["fields"].get("dlibre")
#     codigo_postal = row["record"]["fields"].get("codpostal")
#     direccion = row["record"]["fields"].get("direccion")



data = api.json()

nombres_centros = []
codigos_postales = []
direcciones = []

for index, row in df_centros_edu.iterrows():
    if "record" in row:
        fields = row["record"].get("fields", {})
        nombres_centros.append(fields.get("dlibre"))
        codigos_postales.append(fields.get("codpostal"))
        direcciones.append(fields.get("direccion"))
    elif "fields" in row:  # Check if 'fields' exists directly at the top level
        nombres_centros.append(row.get("dlibre"))
        codigos_postales.append(row.get("codpostal"))
        direcciones.append(row.get("direccion"))
    else:
        nombres_centros.append(None)
        codigos_postales.append(None)
        direcciones.append(None)

df_final = pd.DataFrame({
    "Nombre": nombres_centros,
    "Codigo Postal": codigos_postales,
    "Direccion": direcciones
})

print(df_final)



```
