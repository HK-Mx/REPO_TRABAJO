```python

def plot_predictions_vs_actual(y_real, y_pred):
    """
    Funcin para graficar los valores reales vs. los valores predichos en una regresin.

    Args:
    y_real (array-like): Valores reales de la variable objetivo.
    y_pred (array-like): Valores predichos de la variable objetivo.
    """
    plt.figure(figsize=(8, 6))
    plt.scatter(y_pred, y_real, alpha=0.5)
    plt.xlabel("Valores Predichos")
    plt.ylabel("Valores Reales")

    # Lnea y=x
    max_value = max(max(y_real), max(y_pred))
    min_value = min(min(y_real), min(y_pred))
    plt.plot([min_value, max_value], [min_value, max_value], 'r')

    plt.title("Comparacin de Valores Reales vs. Predichos")
    plt.show()
```

### #1 Explicacin del dataset y carga de datos

Vamos a trabajar con el dataset "Wine Quality", que es un dataset en el que se recogen diversas caractersticas fisicoqumicas de las variades tinta y blanca del "Vinho verde" portugus. Las variables del dataset se describen a continuacin


1. **Acidez fija**: La mayora de los cidos involucrados con el vino son fijos o no voltiles (no se evaporan fcilmente).
2. **Acidez voltil**: La cantidad de cido actico en el vino, que en niveles demasiado altos puede llevar a un sabor desagradable a vinagre.
3. **cido ctrico**: Encontrado en pequeas cantidades, el cido ctrico puede aadir 'frescura' y sabor a los vinos.
4. **Azcar residual**: Es la cantidad de azcar restante despus de que se detiene la fermentacin, es raro encontrar vinos con menos de 1 gramo/litro y vinos con ms de 45 gramos/litro se consideran dulces.
5. **Cloruros**: La cantidad de sal en el vino.
6. **Dixido de azufre libre**: La porcin libre de SO2 existe en equilibrio entre el SO2 molecular (como un gas disuelto) y el ion bisulfito; previene el crecimiento microbiano y la oxidacin del vino.
7. **Dixido de azufre total**: Es la cantidad de formas libres y ligadas de S02; en bajas concentraciones, el SO2 es en su mayora indetectable en el vino, pero en concentraciones libres de ms de 50 ppm, el SO2 se hace evidente en la nariz y el sabor del vino.
8. **Densidad**: La densidad del vino es cercana a la del agua dependiendo del porcentaje de alcohol y contenido de azcar.
9. **pH**: Describe qu tan cido o bsico es un vino en una escala de 0 (muy cido) a 14 (muy bsico); la mayora de los vinos estn entre 3-4 en la escala de pH.
10. **Sulfatos**: Un aditivo del vino que puede contribuir a los niveles de dixido de azufre (SO2), que acta como un antimicrobiano y antioxidante.
11. **Alcohol**: El porcentaje de contenido de alcohol del vino.
12. **Calidad**: Una variable de salida (basada en datos sensoriales, puntuacin entre 0 y 10).
13. **Class**: Si es blanco o tinto.

```python
train_set_scaled = train_set.copy()
test_set_scaled = test_set.copy()
```

```python
features_to_transform = ["residual sugar", "chlorides","free sulfur dioxide", "total sulfur dioxide"]
```

```python
for col in features_to_transform:
    desplaza = 0 
    if train_set_scaled[col].min() <= 0:
        desplaza = int(abs(train_set_scaled[col].min())) + 1
    train_set_scaled[col] = np.log(train_set_scaled[col] + desplaza)
    test_set_scaled[col] = np.log(test_set_scaled[col] + desplaza)
```

```python
features_num_clf
```

*En features_num_clf estn todas las numricas, as que me ahora tener que tratarlas por separado, luego s que las seleccionaremos por separado. Pero ya separando para regresion y para clasificacion*

```python
scaler = StandardScaler()
features_num_reg = features_num_clf.copy()
features_num_reg.remove("alcohol")
train_set_scaled[features_num_reg] = scaler.fit_transform(train_set_scaled[features_num_reg])
test_set_scaled[features_num_reg] = scaler.transform(test_set_scaled[features_num_reg])
```

*Separamos*

```python
train_set_scaled_clf = train_set_scaled.copy()
test_set_scaled_clf = test_set_scaled.copy()

train_set_scaled_clf["alcohol"] = scaler.fit_transform(train_set_scaled_clf[["alcohol"]])
test_set_scaled_clf["alcohol"] = scaler.transform(test_set_scaled_clf[["alcohol"]])
```

*No, no nos hemos olvidado de quality pero ya est convertida a ordinal como numrica discreta que es con rango limitad o pequeo de valores*

*Creamos X,y variados*

```python
y_train_clf = train_set[target_clf]
y_test_clf = test_set[target_clf]
y_train_reg = train_set[target_reg]
y_test_reg = test_set[target_reg]


X_train_clf_dict = {}
X_test_clf_dict = {}

X_train_reg_dict = {}
X_test_reg_dict = {}

# Juegos X_train clasificacion

features_set_name = ["cat+num_1","todas"]
features_set = [features_cat_clf + featues_num_clf_1, features_cat_clf + featues_num_clf_1 + features_num_clf_2]

for set_name,feature_list in zip(features_set_name, features_set):
    print(feature_list)
    X_train_clf_dict[set_name] = train_set_scaled_clf[feature_list]
    X_test_clf_dict[set_name] = test_set_scaled_clf[feature_list]

# Juegos X_train regresion
    
features_set_name = ["cat+num_1","todas"]
features_set = [features_cat_reg + features_num_reg_1, features_cat_reg + features_num_reg_1 + features_num_reg_2]

for set_name,feature_list in zip(features_set_name, features_set):
    X_train_reg_dict[set_name] = train_set_scaled[feature_list]
    X_test_reg_dict[set_name] = test_set_scaled[feature_list]
```

*Ya lo que queda de seccin nos centramos en clasificacin*

```python
# Instanciamos modelos
knn_clf = KNeighborsClassifier(n_neighbors= 5)
knn_ten_clf = KNeighborsClassifier(n_neighbors= 10)
lr_clf = LogisticRegression(max_iter = 1000)
rf_clf = RandomForestClassifier(max_depth = 10, random_state= 42)
lgb_clf = LGBMClassifier(max_depth = 10, random_state = 42, verbose = -100)

modelos_clf = {
    "KNN_5": knn_clf,
    "KNN_10": knn_ten_clf,
    "Regresion Logistica": lr_clf,
    "Random Forest": rf_clf,
    "LightGBM": lgb_clf
}
```

```python
for feature_set, X_train in X_train_clf_dict.items():
    print(f"Para el set {feature_set}:")
    for tipo,modelo in modelos_clf.items():
        print(f"{tipo}: ", end = " ")
        print(np.mean(cross_val_score(modelo, X_train, y_train_clf, cv = 5, scoring = "balanced_accuracy")))
    print("******")
```

*Los valores son realmente flojos, pero es debido al fuerte desequilibrio que hay entre clases, deberamos hacer un equilibrado, pero que en nuestro caso se va a centrar por tiempo disponible en usar el parmetro class_weights y el ms o menos equivalente de weights en KNN*

```python
# Instanciamos modelos
knn_clf = KNeighborsClassifier(n_neighbors= 5, weights= "distance")
knn_ten_clf = KNeighborsClassifier(n_neighbors= 10, weights = "distance")
lr_clf = LogisticRegression(max_iter = 1000, class_weight = "balanced")
rf_clf = RandomForestClassifier(max_depth = 10, random_state= 42, class_weight = "balanced")
lgb_clf = LGBMClassifier(max_depth = 10, random_state = 42, verbose = -100, class_weight = "balanced")

modelos_clf = {
    "KNN_5": knn_clf,
    "KNN_10": knn_ten_clf,
    "Regresion Logistica": lr_clf,
    "Random Forest": rf_clf,
    "LightGBM": lgb_clf
}
```

```python
for feature_set, X_train in X_train_clf_dict.items():
    print(f"Para el set {feature_set}:")
    for tipo,modelo in modelos_clf.items():
        print(f"{tipo}: ", end = " ")
        print(np.mean(cross_val_score(modelo, X_train, y_train_clf, cv = 5, scoring = "balanced_accuracy")))
    print("******")
```

*Todos los modelos han mejorado, e incluso podramos quedarnos con la regresin logistica, pero vamos a hacerlo con LightGBM que parece ms consistente en todas las pruebas y tambin permite a titulo ilustrativo ms juego en el aspecto de hiperparmetros*

```python
param_grid = {
    "n_estimators": [100,200,1000],
    "max_depth": [1,5,10,20],
    "learning_rate": [0.1,0.3,0.6,1,0.01]
}

lgb_grid = GridSearchCV(lgb_clf,
                        param_grid= param_grid,
                        cv = 5,
                        scoring = "balanced_accuracy",
                        n_jobs=-1)

lgb_grid.fit(X_train_clf_dict["todas"], y_train_clf)
```

```python
print(lgb_grid.best_params_)
print(lgb_grid.best_score_)
```

*Evaluamos contra test y veamos la matriz de confusin para entender los errores*

```python
y_pred = lgb_grid.best_estimator_.predict(X_test_clf_dict["todas"])
print(classification_report(y_test_clf, y_pred))
```

```python
ConfusionMatrixDisplay.from_predictions(y_test_clf, y_pred, normalize = None)
```

#### Eleccin entre pickle y joblib

- Utiliza pickle si ests trabajando con modelos simples o quieres apegarte a la biblioteca estndar sin instalar paquetes adicionales.
- Prefiere joblib para guardar y cargar modelos de scikit-learn, especialmente cuando el modelo involucra grandes arreglos de datos. joblib est optimizado para estos escenarios y puede reducir significativamente el espacio de almacenamiento y el tiempo de serializacin.

Ambos mtodos son ampliamente utilizados y soportados para la mayora de los modelos en scikit-learn. Sin embargo, es esencial ser consciente de que, con cualquier mtodo de serializacin, solo debes cargar modelos de fuentes confiables para evitar riesgos de seguridad.
